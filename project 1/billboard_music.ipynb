```python
#converting data type to date
bb_music['date.entered'] = pd.to_datetime(bb_music['date.entered'])
```

```python
bb_music['date.peaked'] = pd.to_datetime(bb_music['date.peaked'])
```

```python
bb_music.head()
```

```python
bb_music = pd.melt(bb_music, id_vars=['artiste', 'track', 'genre', 'date.entered', 'date.peaked'], var_name='week', value_name='rank')
```

```python
bb_music.sample(4)
```

```python
#extracting number of days music entered into peak from date entered and date peaked
bb_music['week'] = bb_music['week'].str.extract('(\d+)').astype(int)
```

```python
bb_music.head()
```

```python
#inserting a new column in data frame (entered_to_peak) = date.peaked - date.entered
bb_music.insert(5,'entered_to_peak', bb_music['date.peaked']-bb_music['date.entered'])
```
