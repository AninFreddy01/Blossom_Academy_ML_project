* Recency – How recently did the customer purchase?
* Frequency – How often do they purchase?
* Monetary Value – How much do they spend?

```python
cut_off_date = df1.InvoiceDate.max() + pd.DateOffset(n=1)
cut_off_date = cut_off_date.normalize()
cut_off_date
```

```python
df1.head()
```

```python
r = df1.groupby('CustomerID', as_index=False).agg({'InvoiceDate': 'max'})
r.head()
```

```python
r['recency'] =  cut_off_date -  r['InvoiceDate'] 
r.head(2)
```

```python
df.InvoiceNo.str
```

```python
r.dtypes
```

```python
r['recency'] = r.recency.dt.round(freq='d')
r.head(2)
```

```python
r['recency'] = r.recency.dt.days
r.head(2)
```
