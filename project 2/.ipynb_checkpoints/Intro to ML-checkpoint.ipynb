Learning happens when the person or machine improves its performance at completing the task. Hence, with some metric, we should be able to tell if there has been an improvement in performing the task.

### What is Machine Learning?
Machine learning is often defined as the ability of a machine to learn (to improve on a specific task) without being explicitly programmed to do so.

## Build a very simple models.

We'll use several ML frameworks eg. scikit-learn. These frameworks automate the learning process for you. Instead of beginning with it, let's develop the habit of using very simple techniques before complex ones. These techniques can even be done on paper.

### What are we predicting?
In this problem, we want to predict the final sale price of the house, ie SalePrice.

### Assign target variable to `y`

```python
y = housing['SalePrice']
y.head()
```

### Baseline model for regression

One of the simplest models we can build for regression problems is guessing the mean.

```python
y_pred = y.mean()
y_pred
```

```python
pd.np.sqrt(180921.19589041095)
```

Here we are saying for every input, we'll use this 'fixed' mean value, 180921.19589041095 as our prediction.

### Assess model performance

For each model we build, we will assess its performance. A popular measure of performance for regression problems is the Mean Squared Error or MSE. This calculation involves three steps:


* Calculate error for each observation; actual - prediction
* Square each error
* Sum the errors
* Divide by the number of observation

#### Step 1: Calculate error

NB: y here is a vector i.e we are subtracting all y values from the mean

```python
error = y - y_pred
error.head()
```

#### Step 2: Square errors

```python
error_squared = error ** 2
error_squared.head()
```

#### Step 3: Sum errors

Get a single metric to objectively quantify our model.

```python
sum_of_errors = error_squared.sum()
sum_of_errors
```

#### Step 4: Compute the mean of the errors

```python
model_1_mse  = sum_of_errors / len(y)
model_1_mse
```

### Model 2 - Use average neighborhood sale price

```python
nbh_mean = housing.groupby('Neighborhood')['SalePrice'].mean()
nbh_mean.head()
```

```python
y_pred = nbh_mean.loc[housing['Neighborhood']].values.round(-3)
y_pred[:10]
```

#### Calculate MSE in one step

```python
model_2_mse = ((y - y_pred) ** 2).sum() / len(y)
model_2_mse
```

### Plot errors for simple models

```python
errors = {
    'model_1_mse': model_1_mse,
    'model_2_mse': model_2_mse, 
}

pd.Series(errors).plot(kind='bar', figsize=(12, 5), rot=0);
```
