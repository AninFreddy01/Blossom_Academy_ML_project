```python
price = df['price']
price
```

```python
price_pred = df['price'].mean()
price_pred
```

```python
error = price - price_pred
error
```

```python
error_squared = error **2
error_squared
```

```python
error_sum = error_squared.sum()
error_sum
```

```python
MSE = error_sum / len(price)
MSE
```

```python
city_mean = df.groupby('city')['price'].mean()
city_mean.head()
```

```python
price_pred = city_mean.loc[df['city']].values.round(-3)
```

```python
MAE = ((price - price_pred)**2).sum() / len(price)
```

```python
MAE
```

```python
errors = {
    'Mean_squared_error': MSE,
    'Mean_absolute_error': MAE, 
}

pd.Series(errors).plot(kind='bar', figsize=(12, 5), rot=0);
```
